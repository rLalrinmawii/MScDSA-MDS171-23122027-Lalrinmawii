## My Menu Driven Restaurant

In the charming little town named "Culinary Haven", there stood a restaurant called "The Enchanted Fork". This restaurant was more than just a place to eat; it was a magical experience that left a lasting impression on all who visited. The restaurant is also famous for its management application that takes up customer orders.

The software collects the order details for the customers. The application generates a unique orderid for each order based on the below specifications. The software system has a predefined menu of items and their prices, from which the user can select them.

OrderID is created with the help of a combination of 5 random numbers generated with the help of a random library in Python. You can use random.randint() function to get one random integer at a time and the order id is generated by concatenating five such numbers.

Apart from orderid, each order will have the items ordered and their quantity saved.

The application has the following options in its menu:

Create new order        
View Orders (List and calculate the bill for each order)       

Assumptions:       
Each order will have a unique orderID generated as per the specifications given.
Each order can have multiple items.        
Once the order is completed, you may calculate the total order value             
You can set a predefined menu and price for the items       

Expectations:
Identify the appropriate data storage mechanisms to store the order data effectively.           
Usage of functions      
menu items stored in the format: [["Rice",30],["Tea",10],.....]             
Each order needs to be saved in the following format           

{
"Orderid": 56837,
"items": [
{
"itemnumber": 2,
"quantity": 1,
},
{
"itemnumber": 1,
"quantity": 1,
},
]
}

```python
import random

# Enter Menu list with prices
menu = [["Chicken Thukpa", 200], ["Veg Thukpa", 150], ["Jajjangmyeon", 250], ["Shrimp tempura maki", 450], ["Chicken Noodle Soup", 180], ["Kimbap", 170]]
orders = []

# Generate random 5-digits orderID (by concatenating)
def random_orderID():
    uniqueOrderID = []
    for i in range(5):
        uniqueOrderID.append(str(random.randint(1, 9))) 
    return "".join(uniqueOrderID)

# Defining a function for new orders
def orderedItems():
    uniqueOrderID = random_orderID()
    # Empty list for items ordered
    Ordered_items = []


    #Print the menu with their prices
    print("MENU:")

    for i, (item, price) in enumerate(menu, 1):
        print(f"{i}.{item} - ₹{price}")

    #enter 0 to stop order
    #enter item number nd quantity to place order
    while True:
        itemNumber = int(input("Enter the item number (enter 0 to finish the order): "))
        if itemNumber == 0:
            break
        if 1 <= itemNumber <= len(menu):
            quantity = int(input("How many do you want to order? "))
            Ordered_items.append({"itemnumber": itemNumber, "quantity": quantity})
        else:
            print("Item number is invalid.")

    


    # Create the order dictionary inside the loop to capture quantity
    order = {"Orderid": uniqueOrderID, "items": Ordered_items}
    orders.append(order)
    print(f"Order {uniqueOrderID} has been confirmed!")
    print("Thank you for your order")
    


# Viewing the orders
def viewOrders():
    for order in orders:
        order_id = order["Orderid"]
        items = order["items"]

        print("Order ID:", order_id)

        # To calculate the total cost of the order
        # Initialize the total cost for this order

        order_total_cost = 0

        for item in items:
            item_number = item["itemnumber"] - 1  #to adjust the indexingas it starts from 0
            item_name, item_price = menu[item_number]
            item_cost = item_price * item["quantity"]

            print(f"{item_name} x {item['quantity']}: ₹{item_cost}")  

            order_total_cost += item_cost 

        print(f"Total Cost: ₹{order_total_cost}\n")  

#looping
# press 1 to create new order
# press 2 to view orders
# press 3 to stop the loop

while True:
    print("1. New Order")
    print("2. View Orders")
    print("3. Quit")
    choice = input("Select an option: ")



    if choice == "1":
        orderedItems()
    elif choice == "2":
        viewOrders()
    elif choice == "3":
        break
    else:
        print("Invalid Entry.")

print("Thank you for choosing The Enchanted Fork!")
```
