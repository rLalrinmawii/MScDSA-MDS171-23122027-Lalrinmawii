```python
# Initialize a list to store information about endangered species
endangered_database = []

# Define a function to add a new endangered species to the database
def add_species(name, population, habitat):
    species = {
        "Name": name,
        "Population": population,
        "Habitat": habitat
    }
    endangered_database.append(species)

# Define a function to update the population of an endangered species
def update_population(name, new_population):
    for species in endangered_database:
        if species["Name"] == name:
            species["Population"] = new_population
            break

# Define a function to calculate the rate of decline or growth in population for an endangered species
def calculate_population_change_rate(name, new_population):
    for species in endangered_database:
        if species["Name"] == name:
            old_population = species["Population"]
            if new_population > old_population:
                change_rate = ((new_population - old_population) / old_population) * 100
                return f"The population of {name} has grown by {change_rate:.2f}%."
            elif new_population < old_population:
                change_rate = ((old_population - new_population) / old_population) * 100
                return f"The population of {name} has declined by {change_rate:.2f}%."
            else:
                return f"The population of {name} has remained unchanged."
    return f"Species {name} not found in the database."

# Define a function to display information about all endangered species in the database
def display_species():
    print("Endangered Species Database:")
    for species in endangered_database:
        print(f"Name: {species['Name']}, Population: {species['Population']}, Habitat: {species['Habitat']}")

# Main program loop
while True:
    print("\nOptions:")
    print("1. Add a new species")
    print("2. Update population of a species")
    print("3. Calculate population change rate")
    print("4. Display all species")
    print("5. Exit")
    
    choice = input("Enter your choice (1/2/3/4/5): ")
    
    if choice == '1':
        name = input("Enter species name: ")
        population = int(input("Enter population count: "))
        habitat = input("Enter habitat type: ")
        add_species(name, population, habitat)
        print(f"{name} added to the database.")
    elif choice == '2':
        name = input("Enter species name to update population: ")
        new_population = int(input("Enter new population count: "))
        update_population(name, new_population)
        print(f"Population of {name} updated.")
    elif choice == '3':
        name = input("Enter species name to calculate population change rate: ")
        new_population = int(input("Enter the current population count: "))
        change_rate_message = calculate_population_change_rate(name, new_population)
        print(change_rate_message)
    elif choice == '4':
        display_species()
    elif choice == '5':
        print("Exiting program. Thank you for your conservation efforts!")
        break
    else:
        print("Invalid choice. Please select a valid option.")
```
