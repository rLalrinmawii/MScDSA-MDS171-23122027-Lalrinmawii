The people of Arithmetica have another issue: they don't believe in the built-in functions (sum, len, max, min, sort, etc.) that Python provides. With the help of functions you design, try to convince the people from Arithmetica that both the outputs (the results generated by your custom functions and the functions that Python Provides) are the same.

```python
data = [761, 123, 585, 275, 194, 840, 999, 622, 763, 40, 825, 687, 402, 338, 611, 987, 739, 95, 819, 568, 320, 126, 688, 948, 514, 223, 751, 379, 857, 767, 492, 162, 437, 934, 430, 343, 725, 511, 580, 678, 708, 823, 346, 179, 786, 663, 195, 890, 267, 502, 214, 300, 786, 961, 235, 559, 992, 919, 948, 318, 57, 787, 653, 317, 305, 172, 966, 182, 970, 17, 89, 624, 629, 595, 485, 752, 984, 978, 460, 124, 321, 256, 614, 852, 811, 561, 127, 207, 406, 615, 888, 391, 691, 470, 911, 401, 137, 600, 478, 682]
```

```python
def Mean_list(data):
    total = 0
    count = 0
    for i in data:
        total += i
        count += 1
    mean = total / count
    return mean

def minimum_list(data):
    minimum = data[0]
    for i in data:
        if i < minimum:
            minimum = i
    return minimum

def maximum_list(data):
    maximum = data[0]
    for i in data:
        if i > maximum:
            maximum = i
    return maximum

def range_list(data):
    minimum = minimum_list(data)
    maximum = maximum_list(data)
    range = maximum - minimum
    return range
```

```python
Mean_list(data)
minimum_list(data)
maximum_list(data)
range_list(data)
print("Mean:", Mean_list(data))
print("Minimum:", minimum_list(data))
print("Maximum:", maximum_list(data))
print("Range:", range_list(data))
```
